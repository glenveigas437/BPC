```python
import pandas as pd
```

```python
league_table=pd.read_csv('QualifierTeams.csv')
goalscorers=pd.read_csv('QualifierGS.csv')
assists=pd.read_csv('QualifiersAs.csv')
cleansheets=pd.read_csv('QualifierCS.csv')
```

```python
from pandas import DataFrame
```

```python
teams=list(league_table['TEAM'])
played=list(league_table['P'])
won=list(league_table['W'])
draw=list(league_table['D'])
lost=list(league_table['L'])
gs=list(league_table['GS'])
gc=list(league_table['GC'])
gd=list(league_table['GD'])
points=list(league_table['pts'])
```

```python
player=list(goalscorers['Player'])
goals=list(goalscorers['Goals'])
team=list(goalscorers['Team'])
```

```python
playm=list(assists['Playmaker'])
ast=list(assists['Assists'])
tm1=list(assists['Team'])
```

```python
keep=list(cleansheets['Keeper'])
cs=list(cleansheets['Clean Sheets'])
tm2=list(cleansheets['Team'])
```

```python
#MATCH
games_played=int(input("HOW MANY GAMES PLAYED THIS MATCHWEEK? "))
for match in range(games_played):
    ht=input("HOME TEAM: ")
    print("GOAL SCORED BY",ht,": ")
    gh=int(input())
    at=input("AWAY TEAM: ")
    print("GOAL SCORED BY",at,": ")
    ga=int(input())
    for t in range(len(teams)):
        if(ht in teams):
            ind_home=teams.index(ht)
            ind_away=teams.index(at)
    
    played[ind_home]=played[ind_home]+1
    played[ind_away]=played[ind_away]+1
    gs[ind_home]=gs[ind_home]+gh
    gc[ind_home]=gc[ind_home]+ga
    gs[ind_away]=gs[ind_away]+ga
    gc[ind_away]=gc[ind_away]+gh
    gd[ind_home]=gs[ind_home]-gc[ind_home]
    gd[ind_away]=gs[ind_away]-gc[ind_away]
    
    
    if(gh>ga):
        points[ind_home]=points[ind_home]+3
        points[ind_away]=points[ind_away]+0
        won[ind_home]=won[ind_home]+1
        lost[ind_away]=lost[ind_away]+1
        
    if(ga>gh):
        points[ind_away]=points[ind_away]+3
        points[ind_home]=points[ind_home]+0
        won[ind_away]=won[ind_away]+1
        lost[ind_home]=lost[ind_home]+1
        
    if(gh==ga):
        points[ind_home]=points[ind_home]+1
        points[ind_away]=points[ind_away]+1
        draw[ind_home]=draw[ind_home]+1
        draw[ind_away]=draw[ind_away]+1    
```

```python
#GOALSCORERS, ASSISTS, CLEANSHEETS
#lgs=len(goalscorers)
#las=len(assists)
#lcs=len(cleansheets)
while(True):
    gsc=input("GOALSCORER: ")
    if(gsc=="X"):
        break
    scored=int(input("NUMBER OF GOALS: "))
    tm=input("TEAM: ")
    #for g in range(lgs):
    if gsc not in player:
        player.append(gsc)
        goals.append(0)
        team.append("A")
    
    if gsc in player:
        player_ind=player.index(gsc)
        goals[player_ind]=goals[player_ind]+scored
        team[player_ind]=tm

print("\n")
print("ASSISTS")
while(True):
    pm=input("ASSIST MAN: ")
    if(pm=="X"):
        break
    assisted=int(input("NUMBER OF ASSISTS: "))
    tmp=input("TEAM: ")
    #for p in range(las):
    if pm not in playm:
        playm.append(pm)
        ast.append(0)
        tm1.append("B")
        #las+=1
    if pm in playm:
        playm_ind=playm.index(pm)
        ast[playm_ind]=ast[playm_ind]+assisted
        tm1[playm_ind]=tmp
            
print("\n")
print("CLEAN SHEETS")
while(True):
    cls=input("KEEPER: ")
    if(cls=="X"):
        break
    css=int(input("NUMBER OF CLEANSHEETS: "))
    tmc=input("TEAM: ")
    #for c in range(lcs):
    if cls not in keep:
        keep.append(cls)
        cs.append(0)
        tm2.append("C")
        keep_ind=keep.index(cls)
        cs[keep_ind]=cs[keep_ind]+css
        tm2[keep_ind]=tmc
        #lcs+=1
    elif cls in keep:
        keep_ind=keep.index(cls)
        cs[keep_ind]=cs[keep_ind]+css
        tm2[keep_ind]=tmc           
```

```python
fin={'TEAM': teams,
     'P': played,
     'W': won,
     'D': draw,
     'L': lost,
     'GS': gs,
     'GC': gc,
     'GD': gd,
     'pts': points
    }

goalscore={'Player': player,
     'Goals': goals,
     'Team': team
    }

asst={'Playmaker': playm,
     'Assists': ast,
     'Team': tm1
    }

cleans={'Keeper': keep,
     'Clean Sheets': cs,
     'Team': tm2
    }
```

```python
df = DataFrame(fin, columns= ['TEAM','P','W','D','L','GS','GC','GD','pts'])
df1 = DataFrame(goalscore, columns= ['Player', 'Goals', 'Team'])
df2 = DataFrame(asst, columns= ['Playmaker', 'Assists', 'Team'])
df3 = DataFrame(cleans, columns= ['Keeper', 'Clean Sheets', 'Team'])
```

```python
df.index +=1
df1.index +=1
df2.index +=1
df3.index +=1
```
