```python
import pandas as pd
```

```python
#league_table=pd.read_csv('TEAMS.csv')
goalscorers=pd.read_csv('Goalscorers.csv')
assists=pd.read_csv('Assists.csv')
cleansheets=pd.read_csv('Cleansheets.csv')
```

```python
from pandas import DataFrame
```

```python
player=list(goalscorers['Player'])
goals=list(goalscorers['Goals'])
team=list(goalscorers['Team'])
```

```python
playm=list(assists['Playmaker'])
ast=list(assists['Assists'])
tm1=list(assists['Team'])
```

```python
keep=list(cleansheets['Keeper'])
cs=list(cleansheets['Clean Sheets'])
tm2=list(cleansheets['Team'])
```

```python
#GOALSCORERS, ASSISTS, CLEANSHEETS
#lgs=len(goalscorers)
#las=len(assists)
#lcs=len(cleansheets)
while(True):
    gsc=input("GOALSCORER: ")
    if(gsc=="X"):
        break
    scored=int(input("NUMBER OF GOALS: "))
    tm=input("TEAM: ")
    #for g in range(lgs):
    if gsc not in player:
        player.append(gsc)
        goals.append(0)
        team.append("A")
    
    if gsc in player:
        player_ind=player.index(gsc)
        goals[player_ind]=goals[player_ind]+scored
        team[player_ind]=tm

print("\n")
print("ASSISTS")
while(True):
    pm=input("ASSIST MAN: ")
    if(pm=="X"):
        break
    assisted=int(input("NUMBER OF ASSISTS: "))
    tmp=input("TEAM: ")
    #for p in range(las):
    if pm not in playm:
        playm.append(pm)
        ast.append(0)
        tm1.append("B")
        #las+=1
    if pm in playm:
        playm_ind=playm.index(pm)
        ast[playm_ind]=ast[playm_ind]+assisted
        tm1[playm_ind]=tmp
            
print("\n")
print("CLEAN SHEETS")
while(True):
    cls=input("KEEPER: ")
    if(cls=="X"):
        break
    css=int(input("NUMBER OF CLEANSHEETS: "))
    tmc=input("TEAM: ")
    #for c in range(lcs):
    if cls not in keep:
        keep.append(cls)
        cs.append(0)
        tm2.append("C")
        keep_ind=keep.index(cls)
        cs[keep_ind]=cs[keep_ind]+css
        tm2[keep_ind]=tmc
        #lcs+=1
    elif cls in keep:
        keep_ind=keep.index(cls)
        cs[keep_ind]=cs[keep_ind]+css
        tm2[keep_ind]=tmc           
```

```python
goalscore={'Player': player,
     'Goals': goals,
     'Team': team
    }

asst={'Playmaker': playm,
     'Assists': ast,
     'Team': tm1
    }

cleans={'Keeper': keep,
     'Clean Sheets': cs,
     'Team': tm2
    }
```

```python
df1 = DataFrame(goalscore, columns= ['Player', 'Goals', 'Team'])
df2 = DataFrame(asst, columns= ['Playmaker', 'Assists', 'Team'])
df3 = DataFrame(cleans, columns= ['Keeper', 'Clean Sheets', 'Team'])
```

```python
df1.index +=1
df2.index +=1
df3.index +=1
```

```python
b=df1.sort_values(by=['Goals', 'Team'], inplace=False, ascending=False)
c=df2.sort_values(by=['Assists'], inplace=False, ascending=False)
d=df3.sort_values(by=['Clean Sheets', 'Keeper'], inplace=False, ascending=False)
```
